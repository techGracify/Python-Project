```python
#  Calculating employees' average salary and their performance rating.

pivot = df.pivot_table(index = "Name", values =["Salary", "Performance Rating"], aggfunc = "mean")
pivot
```

```python
# Sorting top10 employees average salary

top_10 = pivot.sort_values(by  ="Performance Rating", ascending = False).head(10)
top_10
```

```python
top_10.plot(kind = "bar", color = 'gray', figsize =(12,6))
plt.title("Top 10 Employees by Average Salary and Performance Rating")
plt.ylabel("Salary")
plt.xlabel("Employee Names")
plt.xticks()
plt.legend(["Average Salary", "Average Performance Rating"])
plt.grid(axis ="y", linestyle = "--", alpha = 0.7)
plt.tight_layout()
plt.show()
```

### The above chart shows that Tammy Smith has the least average salary while Tara Mccoy has the highest average salary amongst the top 10 most paid employee, the reason could be due to their years of experiece and or their Departments.

```python
# Calculating the average Salaries by Department using the group by method

dept =df.groupby('Department')['Salary'].mean()

dept
```

```python
plt.plot(dept, marker= 'o')
plt.xlabel("Department")
plt.ylabel("Average Salary")
plt.title("Employees' average salary by Department")
plt.show()
```

### The above chart shows the average salary by department and the sales department has the lowest average salary while the Hr department has the highest average salary, the reason could be that the sales department are paid based on their monthly sales performance.

```python
top10_oldest = df.sort_values(by ="Years of Experience", ascending =False).head(10)
top10_oldest
```

```python
pivot_table  =top10_oldest.pivot_table(index = "Name",  values =[ "Salary", "Years of Experience"], aggfunc="max")

pivot_table
```

```python
plt.figure(figsize =(12,8))
sns.heatmap(pivot_table, annot =True,fmt = ".0f", cmap = "YlGnBu", linewidths = 0.5,)
plt.title ("Top 10 oldest Employees by Years of Experience", fontsize = 16)
plt.xlabel ("Years of Experience", fontsize = 12)
plt.ylabel ("Name", fontsize = 12)
plt.show()
```

### The above heatmap chart shows that all the oldest top10 employees have the same years of experience but different ranges of salary this could be because of their respective departments.

```python
# Average Salary by Gender

Gender_sal = df.groupby('Gender')['Salary'].mean()
Gender_sal
```
